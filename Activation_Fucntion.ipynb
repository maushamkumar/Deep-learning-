```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.datasets import make_moons

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
X, y = make_moons(n_samples=100, noise=0.25, random_state=2)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

First I built a neural network with any activation function. just using linear function

```python
model = Sequential()
model.add(Dense(128, input_shape=(2,), activation='linear')) # Here we used linear activation function
model.add(Dense(128, activation='linear'))
model.add(Dense(1, activation='sigmoid'))

adam = Adam(learning_rate=0.01)
model.compile(optimizer=adam, loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(X, y, epochs=500, validation_split=0.2, verbose=0)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X, y.astype(int), clf=model, legend=2)
plt.xlim(-0.7, 0.5)
plt.ylim(-0.8, 0.8)
plt.show()
```

You can see here This neural network is only able to draw a single line.

## Let's use any Activation function instead of linear function

Create another model

```python
model = Sequential()
model.add(Dense(128, input_shape=(2,), activation='relu')) # Here we used linear activation function
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

adam = Adam(learning_rate=0.01)
model.compile(optimizer=adam, loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(X, y, epochs=500, validation_split=0.2, verbose=0)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X, y.astype(int), clf=model, legend=2)
plt.xlim(-0.7, 0.5)
plt.ylim(-0.8, 0.8)
plt.show()
```
